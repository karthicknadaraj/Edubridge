# DATA VISUALISATION

# ARTIST NAME VS POPULARITY IN SPOTIFY

```python
plt.figure(figsize=(14,7))
data.groupby('Artist.Name')['Popularity'].sum().plot.bar(color=['r'])
plt.xlabel("Artist.Name",fontsize=15)
plt.title('Artist Name vs Popularity in Spotify',fontsize=20)
plt.xticks(rotation=90)
plt.ylabel('Popularity',fontsize=15)
plt.grid(True)
plt.show()
```

# TRACK NAME VS POPULARITY IN SPOTIFY

```python
plt.figure(figsize=(14,7))
sns.lineplot(x='Track.Name',y='Popularity',data=data)
plt.xlabel('Track Name',fontsize=15)
plt.ylabel('Popularity',fontsize=15)
plt.title('Track Name vs Popularity in Spotify',fontsize=20)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

# GENRE VALUE COUNTS IN SPOTIFY

```python
plt.figure(figsize=(14,7))
data['Genre'].value_counts().plot.barh(color=['g'])
plt.show()
```

# ARTIST NAME VS BEATS PER MIN

```python
plt.figure(figsize=(14,7))
sns.swarmplot(x='Artist.Name',y='Beats.Per.Minute',data=data)
sns.set_style('darkgrid')
plt.xlabel('Artist Name',fontsize=15)
plt.ylabel('BPM',fontsize=15)
plt.xticks(rotation=90)
plt.title("Artist Name vs BPM ")
plt.grid(True)
plt.show()
```

# TRACK NAME VS LENGTH

```python
plt.figure(figsize=(14,7))
plt.bar(data['Track.Name'],data['Length.'],color=['orange','red'])
plt.xlabel('Track Name',fontsize=15)
plt.xticks(rotation=90)
plt.ylabel('Length',fontsize=15)
plt.title('TRACK NAME vs LENGTH')
plt.show()
```

# ENERGY vs LOUDNESS LEVEL

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x='Loudness..dB..',y='Energy',data=data,color=['red'])
plt.xlabel('Loudness',fontsize=15)
plt.ylabel('Energy',fontsize=15)
plt.title('LOUDNESS VS ENERGY LEVEL')
plt.legend=True
plt.show()
```

# LIVENESS VS POPULARITY

```python
plt.figure(figsize=(18,9))
sns.catplot(x='Liveness',y='Popularity',hue='Artist.Name',data=data)
sns.set_style('whitegrid')
plt.xlabel("LIVENESS",fontsize=15)
plt.ylabel('POPULARITY',fontsize=15)
plt.title('LIVENESS vs POPULARITY')
plt.show()
```
