```python
from sklearn.model_selection import train_test_split
```

```python
x_train,y_train,x_test,y_test=train_test_split(x,y,test_size=0.80,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
l=LinearRegression()
l.fit(x_train,x_test)
```

```python
Predictions=l.predict(y_train)
```

```python
Predictions
```

```python
df=pd.DataFrame({'Actual':y_test,'Predicted':Predictions})
df
```
