# DEFINE X AND Y

```python
d.columns
```

```python
x=d[['Area Income','House Age','No of Rooms','No of Bedrooms','Area Population']]
y=d['Price']
```

```python
x
```

```python
y
```

# SPLIT THE DATA IN TRAINING AND TESTING THE DATA

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
```

```python
print(x_train.shape)
```

```python
print(y_train.shape)
```

```python
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
l=LinearRegression()
l.fit(x_train,y_train)
print("train test complete")
```

# PREDICTING THE TEST RESULT

```python
y_pred=l.predict(x_test)
y_pred
```

```python
l.predict([[79545.45857,5.682861,7.009188,4.09,23086.80050]])
```

```python
d1=pd.DataFrame({'Actual':y_test,'Predict':y_pred})
d1
```

# EVALUATING THE MODEL

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

The value 0.9 to 1 so it is linear

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,y_pred)
```

# SCATTERED PLOT OF ACTUAL AND PREDICTED 

```python
plt.scatter(y_test,y_pred,color='r')
plt.title('Actual vs Predicted')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```

```python
sns.regplot(x=y_test,y=y_pred,color="green")
plt.title("Regressionplot",size=15)
plt.xlabel("Actual value",size=12)
plt.ylabel(" Predicted value",size=12)
plt.show()
```
