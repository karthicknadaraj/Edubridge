```python
plt.figure(figsize=(18,9))
plt.bar(data['Track.Name'],data['Popularity'],color=['y','b','r'])
plt.xlabel("TRACK NAME")
plt.ylabel('POPULARITY')
plt.title('TRACK NAME vs POPULARITY')
plt.xticks(rotation='vertical')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16,8))
data['Artist.Name'].value_counts().plot.bar()
plt.title('ARTIST NAME')
plt.ylabel('quantity')
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.swarmplot(x='Genre',y='Popularity',data=data)
swarmplot.set_title('Genre vs Popularity')
plt.xticks(rotation='vertical')
plt.show()
```
